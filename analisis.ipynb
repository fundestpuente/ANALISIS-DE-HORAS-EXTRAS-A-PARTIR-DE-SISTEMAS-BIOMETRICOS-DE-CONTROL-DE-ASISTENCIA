```python
# Gráficas
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Configurar el estilo de las gráficas
#import scienceplots
#plt.style.use(['science', 'notebook', 'no-latex'])

plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Agregar columna de mes a cada dataframe
rep_marzo['Mes'] = 'Marzo'
rep_abril['Mes'] = 'Abril'
rep_mayo['Mes'] = 'Mayo'
rep_junio['Mes'] = 'Junio'

# Combinar todos los datos
todos_meses = pd.concat([rep_marzo, rep_abril, rep_mayo, rep_junio], ignore_index=True)

# Calcular costo (asumiendo $10 por hora extra)
costo_hora_extra = 10
todos_meses['Costo Horas Extras'] = todos_meses['Total horas extras 50%'] * costo_hora_extra

# 1. GRÁFICA COMPARATIVA: Evolución mensual del total de horas extras
plt.figure(figsize=(15, 12))

# Evolución del total de horas extras por mes
plt.subplot(2, 2, 1)
total_por_mes = todos_meses.groupby('Mes')['Total horas extras 50%'].sum()
meses = ['Marzo', 'Abril', 'Mayo', 'Junio']
total_por_mes = total_por_mes.reindex(meses)  # Asegurar orden correcto

plt.plot(meses, total_por_mes, marker='o', linewidth=2, markersize=8, color='blue')
plt.fill_between(meses, total_por_mes, alpha=0.3, color='blue')
plt.title('Evolución de Horas Extras por Mes', fontsize=14, fontweight='bold')
plt.ylabel('Total Horas Extras 50%')
plt.grid(True, alpha=0.3)

# Añadir valores en los puntos
for i, valor in enumerate(total_por_mes):
    plt.text(i, valor + 5, f'{valor:.1f}h', ha='center', va='bottom')

# 2. GRÁFICA: Evolución del costo mensual
plt.subplot(2, 2, 2)
costo_por_mes = todos_meses.groupby('Mes')['Costo Horas Extras'].sum()
costo_por_mes = costo_por_mes.reindex(meses)

plt.bar(meses, costo_por_mes, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4'])
plt.title('Evolución del Costo en Horas Extras por Mes', fontsize=14, fontweight='bold')
plt.ylabel('Costo Total ($)')

# Añadir valores en las barras
for i, valor in enumerate(costo_por_mes):
    plt.text(i, valor + 10, f'${valor:.2f}', ha='center', va='bottom')

# 3. GRÁFICA: Top 5 empleados con más horas extras en todos los meses
plt.subplot(2, 2, 3)
top_5_global = todos_meses.groupby('Nombre')['Total horas extras 50%'].sum().nlargest(5)

plt.barh(top_5_global.index, top_5_global.values, color='lightcoral')
plt.xlabel('Total Horas Extras Acumuladas')
plt.title('Top 5 Empleados - Más Horas Extras (Acumulado 4 Meses)', fontsize=12, fontweight='bold')

# Añadir valores en las barras
for i, valor in enumerate(top_5_global.values):
    plt.text(valor + 1, i, f'{valor:.1f}h', va='center')

# 4. GRÁFICA: Distribución de empleados que hacen horas extras por mes
plt.subplot(2, 2, 4)
empleados_con_he_por_mes = []
for mes in meses:
    df_mes = todos_meses[todos_meses['Mes'] == mes]
    empleados_con_he = len(df_mes[df_mes['Total horas extras 50%'] > 0])
    empleados_con_he_por_mes.append(empleados_con_he)

plt.bar(meses, empleados_con_he_por_mes, color='lightgreen')
plt.title('Empleados con Horas Extras por Mes', fontsize=14, fontweight='bold')
plt.ylabel('Número de Empleados')

# Añadir valores en las barras
for i, valor in enumerate(empleados_con_he_por_mes):
    plt.text(i, valor + 0.1, f'{valor}', ha='center', va='bottom')

plt.tight_layout()
plt.show()

# 5. GRÁFICA: Comparación de los 5 empleados con más horas extras entre meses
empleados_top = top_5_global.index.tolist()

plt.figure(figsize=(15, 10))
for i, empleado in enumerate(empleados_top, 1):
    plt.subplot(2, 3, i)
    datos_empleado = todos_meses[todos_meses['Nombre'] == empleado]
    
    # Asegurar que tenemos datos para todos los meses
    datos_completos = []
    for mes in meses:
        dato_mes = datos_empleado[datos_empleado['Mes'] == mes]
        if len(dato_mes) > 0:
            datos_completos.append(dato_mes['Total horas extras 50%'].values[0])
        else:
            datos_completos.append(0)
    
    plt.plot(meses, datos_completos, marker='s', linewidth=2, markersize=6)
    plt.title(f'{empleado}', fontweight='bold')
    plt.ylabel('Horas Extras')
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3)
    
    # Añadir valores en los puntos
    for j, valor in enumerate(datos_completos):
        plt.text(j, valor + 1, f'{valor:.1f}h', ha='center', va='bottom')

plt.tight_layout()
plt.show()

# 6. GRÁFICA: Heatmap de horas extras por empleado y mes
plt.figure(figsize=(16, 10))

# Crear matriz para el heatmap
empleados_unicos = todos_meses['Nombre'].unique()
heatmap_data = []

for empleado in empleados_top:  # Usamos solo los top 5 para mejor visualización
    fila = []
    for mes in meses:
        dato = todos_meses[(todos_meses['Nombre'] == empleado) & (todos_meses['Mes'] == mes)]
        if len(dato) > 0:
            fila.append(dato['Total horas extras 50%'].values[0])
        else:
            fila.append(0)
    heatmap_data.append(fila)

heatmap_data = np.array(heatmap_data)

# Crear heatmap
plt.imshow(heatmap_data, cmap='YlOrRd', aspect='auto')
plt.colorbar(label='Horas Extras')
plt.xticks(range(len(meses)), meses)
plt.yticks(range(len(empleados_top)), empleados_top)
plt.xlabel('Mes')
plt.ylabel('Empleado')
plt.title('Heatmap: Horas Extras por Empleado y Mes', fontsize=14, fontweight='bold')

# Añadir valores en las celdas
for i in range(len(empleados_top)):
    for j in range(len(meses)):
        plt.text(j, i, f'{heatmap_data[i, j]:.1f}', 
                ha='center', va='center', 
                color='black' if heatmap_data[i, j] < heatmap_data.max()/2 else 'white')

plt.tight_layout()
plt.show()

# 7. GRÁFICA: Tendencia de los 3 empleados con más horas extras
plt.figure(figsize=(12, 8))

top_3_empleados = top_5_global.head(3).index.tolist()

for empleado in top_3_empleados:
    datos_empleado = todos_meses[todos_meses['Nombre'] == empleado]
    datos_por_mes = []
    
    for mes in meses:
        dato_mes = datos_empleado[datos_empleado['Mes'] == mes]
        if len(dato_mes) > 0:
            datos_por_mes.append(dato_mes['Total horas extras 50%'].values[0])
        else:
            datos_por_mes.append(0)
    
    plt.plot(meses, datos_por_mes, marker='o', linewidth=3, markersize=8, label=empleado)

plt.title('Tendencia de los 3 Empleados con Más Horas Extras', fontsize=14, fontweight='bold')
plt.ylabel('Horas Extras 50%')
plt.xlabel('Mes')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# RESUMEN ESTADÍSTICO COMPARATIVO
print("="*60)
print("ANÁLISIS COMPARATIVO - MARZO A JUNIO 2024")
print("="*60)

for mes in meses:
    df_mes = todos_meses[todos_meses['Mes'] == mes]
    total_horas = df_mes['Total horas extras 50%'].sum()
    total_costo = df_mes['Costo Horas Extras'].sum()
    empleados_con_he = len(df_mes[df_mes['Total horas extras 50%'] > 0])
    
    print(f"\n{mes}:")
    print(f"  • Total horas extras: {total_horas:.1f} h")
    print(f"  • Costo total: ${total_costo:.2f}")
    print(f"  • Empleados con HE: {empleados_con_he}/{len(df_mes)}")
    print(f"  • Promedio HE por empleado: {total_horas/len(df_mes):.1f} h")

# Análisis del empleado más consistente
empleado_consistente = todos_meses.groupby('Nombre')['Total horas extras 50%'].mean().idxmax()
promedio_consistente = todos_meses.groupby('Nombre')['Total horas extras 50%'].mean().max()

print(f"\nEMPLEADO MÁS CONSISTENTE:")
print(f"  • {empleado_consistente} - Promedio: {promedio_consistente:.1f} h/mes")

# Costo acumulado de 4 meses
costo_total_4_meses = todos_meses['Costo Horas Extras'].sum()
print(f"\nINVERSIÓN TOTAL 4 MESES: ${costo_total_4_meses:.2f}")
```
